```python
import pandas as pd
data = pd.read_csv('C:/Users/ratho/survey lung cancer.csv')
data
```

```python
# filtering the dataset

data['GENDER'] = data['GENDER'].replace('M',1)
data['GENDER'] = data['GENDER'].replace('F',0)
data['SMOKING'] = data['SMOKING'].replace(1,0)
data['SMOKING'] = data['SMOKING'].replace(2,1)



data['LUNG_CANCER'] = data['LUNG_CANCER'].replace('YES',1)
data['LUNG_CANCER'] = data['LUNG_CANCER'].replace('NO',0)

data.head()
```

```python
data['YELLOW_FINGERS'] = data['YELLOW_FINGERS'].replace(1,0)
data['YELLOW_FINGERS'] = data['YELLOW_FINGERS'].replace(2,1)
data.head()
```

```python
data['ANXIETY'] = data['ANXIETY'].replace(1,0)
data['ANXIETY'] = data['ANXIETY'].replace(2,1)
data.head()
```

```python
data['PEER_PRESSURE'] = data['PEER_PRESSURE'].replace(1,0)
data['PEER_PRESSURE'] = data['PEER_PRESSURE'].replace(2,1)
data.head()
```

```python
data['CHRONIC DISEASE'] = data['CHRONIC DISEASE'].replace(1,0)
data['CHRONIC DISEASE'] = data['CHRONIC DISEASE'].replace(2,1)
data.head()
```

```python
data['FATIGUE '] = data['FATIGUE '].replace(1,0)
data['FATIGUE '] = data['FATIGUE '].replace(2,1)
data.head()
```

```python
data['ALLERGY '] = data['ALLERGY '].replace(1,0)
data['ALLERGY '] = data['ALLERGY '].replace(2,1)
data.head()
```

```python
data['WHEEZING'] = data['WHEEZING'].replace(1,0)
data['WHEEZING'] = data['WHEEZING'].replace(2,1)
data.head()
```

```python
data['ALCOHOL CONSUMING'] = data['ALCOHOL CONSUMING'].replace(1,0)
data['ALCOHOL CONSUMING'] = data['ALCOHOL CONSUMING'].replace(2,1)
data.head()
```

```python
data['COUGHING'] = data['COUGHING'].replace(1,0)
data['COUGHING'] = data['COUGHING'].replace(2,1)
data.head()
```

```python
data['SHORTNESS OF BREATH'] = data['SHORTNESS OF BREATH'].replace(1,0)
data['SHORTNESS OF BREATH'] = data['SHORTNESS OF BREATH'].replace(2,1)
data.head()
```

```python
data['SWALLOWING DIFFICULTY'] = data['SWALLOWING DIFFICULTY'].replace(1,0)
data['SWALLOWING DIFFICULTY'] = data['SWALLOWING DIFFICULTY'].replace(2,1)
data.head()
```

```python
data['CHEST PAIN'] = data['CHEST PAIN'].replace(1,0)
data['CHEST PAIN'] = data['CHEST PAIN'].replace(2,1)
data.head()
```

```python
data = data.dropna()
```

```python
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
```

```python
data = data[data['GENDER']<=1].copy()
data = data[data['AGE']<=78].copy()
data = data[data['SMOKING']<2].copy()
data = data[data['YELLOW_FINGERS']<2].copy()
data = data[data['ANXIETY']<2].copy()
data = data[data['PEER_PRESSURE']<2].copy()
data = data[data['CHRONIC DISEASE']<2].copy()
data = data[data['FATIGUE ']<2].copy()
data = data[data['CHEST PAIN']<2].copy()
data = data[data['WHEEZING']<2].copy()
data = data[data['ALCOHOL CONSUMING']<2].copy()
data = data[data['SHORTNESS OF BREATH']<2].copy()
data = data[data['SWALLOWING DIFFICULTY']<2].copy()
data = data[data['LUNG_CANCER']<1].copy()
```

```python
y = data['LUNG_CANCER']
data = data.drop(columns=['LUNG_CANCER'])
data.head()
```

```python
# mod = KNeighborsRegressor().fit(data,y)
# pred = mod.predict(data)
# plt.scatter(pred,y)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state = 42)

x_res , y_res = sm.fit_resample(data,y)
x_res = pd.DataFrame(x_res)
y_res = pd.DataFrame(y_res)

y_res.hist()
```

```python
import matplotlib.pyplot as plt
plt.hist(y_res)
plt.show()
```

```python
# # !pip install sklearn
# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()
# x_res=scaler.fit_transform(x_res)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_res,y_res,test_size=0.24,random_state=111)
```

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier(random_state=12)
sgd.fit(x_train,y_train)
sgd.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=31)
neigh.fit(x_train,y_train)
neigh.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier as DT
DTmodel = DT(max_depth=1)
DTmodel.fit(x_train, y_train)
DTmodel.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 31)

rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier
# fit model no training data

model = XGBClassifier(n_estimators=42)
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
import pickle
import numpy as np
filename = 'lung_cancer_normalized.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
# comparesion chart
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
models = ['xgboost','RandomForest','KNeighborsClassifier','DecisionTree','KNeighbors']
score = [model.score(x_test,y_test)*100,rf.score(x_test,y_test)*100,neigh.score(x_test,y_test)*100,DTmodel.score(x_test,y_test)*100,neigh.score(x_test,y_test)*100]
#print(score)
plt.bar(models,score)
plt.show()
```
